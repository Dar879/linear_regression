```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import data
salary=pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv')
salary.columns
```

```python
#define target(y) and features (X)
y=list(salary['Salary'])
X=list(salary['Experience Years'])
```

```python
#add the bias terms
X_ones=np.c_[np.ones(40),X]
X_ones
```

```python
#apply normal equation
theta= np.linalg.inv(X_ones.T.dot(X_ones)).dot(X_ones.T).dot(y)
theta
```

```python
#draw points 
plt.scatter(X,y,s=4, c='#07d6a2')

#draw line
X_lim=[0,11]
X_lim_ones=np.c_[np.ones(2),X_lim]
y_lim=X_lim_ones.dot(theta)
plt.plot(X_lim,y_lim,'r-')

#grahp config

plt.axis([0,11,20000,140000])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Experience Years x Salary')
plt.grid()
```

```python
#Predict a Salary
salary_Lucas=theta[0]+theta[1] * 7.5
salary_Lucas
```
